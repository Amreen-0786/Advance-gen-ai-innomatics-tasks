# Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = 0

for name in employees:
    if employees[name] > highest_score:
        highest_score = employees[name]

print("Top Performers Eligible for Bonus:")

for name in employees:
    if employees[name] == highest_score:
        print(name)

print("Score:", highest_score)
```

# Search Query Keyword Analysis

```python
query = "Buy mobile phone buy phone online"

query = query.lower()
words = query.split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] = frequency[word] + 1
    else:
        frequency[word] = 1

print("Repeated Keywords:")

for word in frequency:
    if frequency[word] > 1:
        print(word, ":", frequency[word])
```

# Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    parts = email.split("@")
    domain = parts[1]

    if domain in domain_count:
        domain_count[domain] = domain_count[domain] + 1
    else:
        domain_count[domain] = 1

total = len(emails)

for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(domain, ":", int(percentage), "%")
```

# Sales Spike Detection

```python
# Problem 5

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
total = 0

for s in sales:
    total = total + s

average = total / len(sales)

# 30% above average
threshold = average * 1.30

# Check spike days
for i in range(len(sales)):
    if sales[i] > threshold:
        print("Day", i+1, ":", sales[i])
```

# Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count_ids = {}

for user in user_ids:
    if user in count_ids:
        count_ids[user] = count_ids[user] + 1
    else:
        count_ids[user] = 1

for user in count_ids:
    if count_ids[user] > 1:
        print(user, "->", count_ids[user], "times")
```
